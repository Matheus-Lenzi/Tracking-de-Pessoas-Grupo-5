1.   Baixar o video




```python
# To clone from Github mirror (may be out of sync):
# !git clone https://github.com/ifzhang/FairMOT.git

# './FairMOT/videos/MOT16-03.mp4'
#download a zip with videos to test
!wget https://motchallenge.net/data/MOT15.zip
!unzip MOT15.zip

#turn imags sequence into a video. Fonte https://stackoverflow.com/questions/44947505/how-to-make-a-movie-out-of-images-in-python
import cv2
import os

image_folder = '/content/MOT15/test/ADL-Rundle-1/img1'
video_name = 'video_testeeeeee.avi'

images = [img for img in os.listdir(image_folder) if img.endswith(".jpg")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

video = cv2.VideoWriter(video_name, 0, 1, (width,height))

for image in images:
    video.write(cv2.imread(os.path.join(image_folder, image)))

cv2.destroyAllWindows()
video.release()
```
